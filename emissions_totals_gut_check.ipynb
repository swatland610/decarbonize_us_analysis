```python
import pandas as pd
pd.set_option('display.max_columns', None)

emissions = pd.read_csv('charts/us_emissions_2000_2018.csv')

sub_sectors = ['emission_sub_electric', 'emission_sub_commercial', 'emission_sub_residential', 
               'emission_sub_industrial', 'emission_sub_transportation', 'emission_sub_fugitive', 
               'emission_by_waste', 'emission_by_agriculture', 'emission_by_industrial']
```

```python
emissions.info()
```

```python
emissions['sub_sector_total'] = emissions.apply(lambda row: row[sub_sectors].sum(), axis=1)
```

```python
emissions['total_sub_difference'] = emissions['total_emissions_excluding_lucf'] - emissions['sub_sector_total']
```

```python
emissions['%_diff'] = round(emissions['total_sub_difference'] / emissions['total_emissions_excluding_lucf'], 3)
```

```python
emissions.groupby('State').agg({'%_diff':'mean'}).sort_values(by='%_diff', ascending=False)
```
